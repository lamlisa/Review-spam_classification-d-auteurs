### Calculs des scores de confiance pour les reviews, reviewers et produits

```python
from data import *
import review_graph as rg

nb_rounds = 4
#delta_t = 1209600 # 2 semaines
delta_t = 3*30*24*60*60  # 3 mois
delta = 1 # systeme de notes à 5 étoiles sur ces données

rel,trust,h = rg.compute(productsID, reviewsID, reviewersID, nb_rounds, products, 
                         reviewers, delta_t, delta, avg_notes, review_author, notes,
                         time_post, review_product)
```

### Spammeurs potentiels

**Personne avec le score de trustiness le plus mauvais : spammeur présumé**

```python
worst_score_reviewer = min(trust, key=trust.get)
print("worst_score_reviewers : {}".format(worst_score_reviewer))
```

**Reviews du présumé spammeur**

```python
df1000[df1000['reviewerID']==worst_score_reviewer]
```

**Score d'honnêteté de ses reviews**

```python
temp = df1000[df1000['reviewerID']==worst_score_reviewer].index
h[temp]
```

Le score d'honnêteté est très mauvais (presque -1) ce qui est cohérent avec un mauvais score de trustiness de l'auteur.

**Reviews d'un des produits qui a reçu un avis du présumé spammeur**

```python
df1000[df1000['asin']=='B000H26EMY']
```

**Moyennne des scores de trustiness par note pour ce produit**

```python
for note in range(1,6):
    temp = df[(df['asin']=='B000H26EMY') & (df['overall']==note)]['reviewerID']
    print("note {} : {}".format(note,sum([trust[i] for i in temp])/len(temp)))
```

```python
#compte le nombre de reviewers pour chaque note
df[df['asin']=='B000H26EMY'].groupby(['overall']).size()
```

Cohérent, le présumé spammeur a mis une note de 1 et on voit qu'en moyenne les auteurs ayant mis une mauvaise note à ce produit ne sont pas fiables (score de trustiness négatif).

**Reviewer avec meilleur score**

```python
best_score_reviewer = max(trust, key=trust.get)
print("best_score_reviewers : {}".format(best_score_reviewer))
```

**Reviews de la personne la plus fiable**

```python
df[df['reviewerID']==best_score_reviewer]
```

**reviews d'un des produits qui a reçu un avis du meilleur reviewer**

```python
#reviews d'un des produits qui a reçu un avis du meilleur reviewer
df[df['asin']=='B000H0X79O']
```

**Score d'honnêteté de ses reviews**

```python
temp = df[df['reviewerID']==best_score_reviewer].index
h[temp]
```

Ses scores d'honnêteté sont élevés, cohérent.

**Moyennne des scores de trustiness par note pour ce produit**

```python
for note in range(1,6):
    temp = df[(df['asin']=='B000H0X79O') & (df['overall']==note)]['reviewerID']
    print("note {} : {}".format(note,sum([trust[i] for i in temp])/len(temp)))
```

```python
#compte le nombre de reviewers pour chaque note
df[df['asin']=='B000H0X79O'].groupby(['overall']).size()
```

Là aussi, cohérence, la personne la plus fiable a mis une note de 5 à ce produit, et la moyenne de trustiness de ceux qui ont mis une bonne note à ce produit est élevée.
